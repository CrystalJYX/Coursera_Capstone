#  Introduction/Business Problem

A description of the problem and a discussion of the background. 

With the frequent occurrence of car accidents, people can summarize the causes of past accidents and ensure safety in the future. However, due to the various causes of car accidents, it is difficult for us to avoid car accidents under special circumstances. Nowadays, we can extract important information from big data to combine the actual situation and give reasonable estimation results to drivers, which could warn drivers and effectively reduce car accidents. Given the weather and the road conditions about the possibility of getting severe into a car accident and how it would be so that people would drive more carefully or even change travels if they are able to. Hence, it is meaningful to build the model and predict the severity of the accident with given conditions.
